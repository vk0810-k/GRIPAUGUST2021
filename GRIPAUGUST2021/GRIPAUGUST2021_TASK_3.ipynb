*Lets import important libraries first!*

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
print("The libraries are imported 100%")
```

*Lets import the dataset now!*

```python
df =  pd.read_csv("SampleSuperstore.csv")
print("The dataset has been imported 100%")
```

*Lets print the first 5 rows now!*

```python
df.head()
```

*Lets check if there is any null value or not!*

```python
df.info()
```

*Lets find the data description of the dataset!*

```python
df.describe()
```

*Lets find the unique data from the dataset!*

```python
df.nunique(axis=0)
```

*Lets calculate the correlation matrix*

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, 
            cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

*Understanding Relationships between categorical data*

```python
sns.catplot(x="Segment",y="Profit",data=df,kind='bar',ci=95)
```

```python
sns.catplot(x="Ship Mode",y="Profit",data=df,kind='bar',ci=None)
```

```python
order = df.groupby(["State"])["Profit"].mean().sort_values().index
sns.catplot(y="State",x="Profit",data=df,kind='bar',ci=None,height=20,order=order)
```

*we observe states like Ohio, North Carolina, Illinois yeild average loss*

```python
sns.catplot(x="Category",y="Profit",data=df,kind="bar",ci=None)
```

*There is significantly high profits in the 'Technology' category and lesser profit in 'Furtniture' category.*

```python
sns.catplot(x="Category",y="Profit",data=df,kind='bar',ci=None,col="Region")
```

*Selling Furtinure in Central region is not advised.*

*Lets do the count plot*

```python
plt.subplots(figsize=(8,10))
order=df.groupby(['State'])["State"].count().sort_values().index
sns.countplot(y='State',data=df,order=order)
```
